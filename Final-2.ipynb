# Overview

Coronavirus disease (COVID-19) is an infectious disease caused by a newly discovered coronavirus. The virus is highly infectious and spreads rapidly through community transmission.  Data regarding COVID-19 changes daily as more tests become available, more citizens are tested, and antibody blood tests are used. The accuracy of the data used for the following visualizations is uncertain.  Lack of testing reduces the reliability of the numbers.

```python
df = pandas.read_excel('s3://alex-ia241/COVID-19-geographic-disbtribution-worldwide-2020-04-28.xlsx')

df
```

## Countries with the Highest Death Toll

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
sum_death_by_country.nlargest(10).plot.bar()
```

The United States far surpasses other countries in terms of COVID-19 related deaths.  Italy and China were the most concerning countries at the beginning of this pandemic, but the US's surge in cases leads to more deaths.  

## China's Cases Per Day  

```python
china_data = df.loc[df['countriesAndTerritories'] == 'China']
china_data[:10]
```

```python
china_data.plot(x="dateRep", y="cases")
```

This visualization goes back to the concern of the reliability of the data from China. There is a steady increase of cases at the end of January to mid February, after which there is a sharp spike in cases.  It is uncertain whether this is accurate or if there were mistakes in the data used.

## Italy's Cases related to Deaths

```python
italy_data = df.loc[df['countriesAndTerritories'] == 'Italy']
italy_data.plot.scatter(x='cases',y='deaths')
```

Italy experienced rapid outbreaks of the virus shortly after concern rose in China.  This visualization shows the relationships between number of cases and number of deaths.  It is clear that there is a strong, positive correlation between the two variables. 

# Conclusion

Limitations include lack of testing in nearly every country, as well as asymptomatic cases that are likely unaccounted for.
